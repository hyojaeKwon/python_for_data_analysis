<h1>NumPy</h1> is a library of Python 

- processing of matrices or large multidimensional arrays. 
- provides implemented function for numerical calculations.

```python
import numpy as np
```

<h2>Numpy Create Array</h2>



```python
#create a rank 1 ndarray
a = np.array([1, 2, 3])
print(type(a))
print(a.shape)
print(a)
```

<h2>List vs Numpy</h2>

1. Array operation: faster in NumPy
2. Functional terms: much more in NumPy
3. NumPy takes up less memory!


<h2>Numpy Index</h2>

Just similar with 'List‘…



```python
b = np.array([[1,2,3],[4,5,6]])   
# Create a rank 2 array

print(b.shape)                    
# Prints "(2, 3)" because it is 2x3 array

print(b[0, 0], b[0, 1], b[0][1: ])   
# Prints "1 2 [2 3]"
```

<h2>Numpy Broadcasting</h2>

This too, array calculation is same with python list!

```python
arr1 = np.array([[1,2,3,4,],[8,9,10,11]])

print(arr1 + arr1)

print() # for the blank space

print(arr1 * arr1)
```

<h2>Numpy Reshape</h2>

It can change the dimension!

```python
arr1 = np.arange(50)
print(arr1)

arr2 = np.arange(50).reshape((10, 5)) # change arr to 10 x 5
print(arr2)

```

<h2>Numpy File I/O</h2>

<b>np.save</b>: make and save array data in disk
<b>np.load</b>: load array data file from disk

```python
arr = np.arange(5)
np.save('my_array',arr) # make my_array.npy 
arr1 = np.load('my_array.npy') # load my_array.npy file 
```
