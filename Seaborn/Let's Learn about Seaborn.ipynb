### rugplot

display the location of the data on the small rug on the x-axis <br>
 http://seaborn.pydata.org/generated/seaborn.rugplot.html

```python
x = iris.petal_length.values

sns.rugplot(x)
plt.title("Rug Plot of the petal_length of iris")
plt.show()
```

### kdeplot -> kernal density

distribution curve that is softer than the histogram.
 http://seaborn.pydata.org/generated/seaborn.kdeplot.html

```python
sns.kdeplot(x)
plt.title("Kernel Density Plot of iris")
plt.show()
```

### distplot
rug + kernal

```python
sns.distplot(x, kde=True, rug=True)
plt.title("Dist Plot of iris flower")
plt.show()
```

## visualize Multi dimensial data

If the data are two-dimensional and all of the data are continue real values, use a <b>scatter</b> plot.



To draw a Scatter Plot, use the <b>jointplot</b>. <br> 
 The jointplot draws a histogram of each variable at the edge of the chart as well as the scatter plot.<br>
 http://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sns.jointplot(x="x_name", y="y_name", data=dataframe, kind='scatter')
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.suptitle("speal length and width in Joint Plot", y=1.02)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.suptitle("speal length and width in Joint Plot Joint Plot , Kernel Density Plot", y=1.02)
plt.show()
```

#### If the data is over 3-dimensial data, use "pairplot" of seaborn
- The pairplot takes the data frame as an argument and draws a scanner plot for each combination of data columns in grid form.
- (if the same data meets) draw a histogram

```python
sns.pairplot(iris)
plt.title("Iris Data Pair Plot")
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.title("Iris Pair Plot, Color - visualize")
plt.show()
# Use "hue" parameter and let it know the name of the variables.
```

## 2dim category data

If all data are 2-dim and all values are category,  USE a heatmap!

```python
titanic_size = titanic.pivot_table(
    index="class", columns="sex", aggfunc="size")

titanic_size
```

```python
sns.heatmap(titanic_size, cmap=sns.light_palette(
    "gray", as_cmap=True), annot=True, fmt="d")
plt.title("Heatmap")
plt.show()
```

## Two-Dimensional Complex Data
If the data are two-dimensional and mixed with real and category valus, you can use these plots <br>
- barplot
- violinplot
- stripplot, swarmplot

### Barplot
Barplot produces a basic bar chart that displays the mean and deviation of real values according to category values. <br>
The mean is the height of the bar and the deviation is indicated by the error bar.<br>
http://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("total bill for each day")
plt.show()
```

### violin plot
It provides the oveall data distribution <br>
Kernel Density histogram with vertical direction.<br>
http://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("total bill for the each day Violin Plot")
plt.show()
```

### strip plot
<b>Draw all data with dots.</b><br>
jitter=True  ==> it does not overlap when the number of data is large.

```python
import numpy as np
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Strip Plot")
plt.show()
```

```python
#Move aside so that the points do not overlap.
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("Swarm Plot")
plt.show()
```

### Multidimensional Complex Data
All functions also include the visualize to analyze high-dimensional data above two dimensions.

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("multidimensional Histogram")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("multidimensional Violin Plot")
plt.show()
```

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True)
plt.title("Multidimensional Strip Plot")
plt.legend(loc=1)
plt.show()
```

## we can visualize with different style with and split"

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, dodge=True)
plt.title("multidimensional Violin Plot, Split=True")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", hue="sex",
              data=tips, jitter=True, dodge=True)
plt.title("multidimensional Strip Plot, Split=True")
plt.show()
```

### you can see the variation about two-categories with heatmap

```python
flights_passengers = flights.pivot("month", "year", "passengers")
plt.title("multi dimensional Heatmap")
sns.heatmap(flights_passengers, annot=True, fmt="d", linewidths=1)
plt.show()

#http://seaborn.pydata.org/generated/seaborn.heatmap.html
```
